#### 🚨 Troubleshooting Tip 1: Text Disappeared into Raw Code?
As you work through this notebook, you might accidentally double-click a Markdown (text) section like this one. If you do, the clean text and instructions will vanish, turning into a box full of raw code symbols, hashtags (#), or bracketed text.
Whenever a text block opens up into raw code, it just means you accidentally entered "Edit Mode." To fix it:

   1. Click your mouse anywhere inside that broken box.
   2. Press **Shift + Enter** on your keyboard.
   3. The computer will instantly re-process the code and return the text to normal.

#### 🔄 Troubleshooting Tip 2: The Universal Reset Button
Sometimes, a typo can cause a code cell to get stuck, loop forever, throw a giant wall of red error text, or freeze your screen entirely.
If your notebook locks up or stops responding to your commands:

   1. Look at the menu bar at the very top of your notebook screen.
   2. Click on the word **Kernel**.
   3. Select **Restart Kernel...** from the dropdown menu.
   4. This acts like a quick reboot. It clears out the frozen code memory and gives you a fresh, clean slate to run your code cells again.

```python
# NOTE: The word Hello is followed by a space and both are placed inside quotation marks. 
# Spaces are also strings in python and will print out along with the word.
# The number 5 is not inside quotation marks.
#
# Click SHIFT+ENTER to run the cell.

print("-----------------------------")
print("Hello " * 5)
print("-----------------------------")
```
