#### 🚨 Troubleshooting Tip 1:

#### 🔄 Troubleshooting Tip 2: The Universal Reset Button

Sometimes, a typo can cause a code cell to get stuck, loop forever, throw a giant wall of red error text, or freeze your screen entirely. 

If your notebook locks up or stops responding to your commands:

1. Look at the very top menu bar of your screen.
2. Click on the word **Runtime** (if you are on Google Colab) or **Kernel** (if you are on a computer desktop).
3. Select **Restart Runtime** (or **Restart Kernel**) from the dropdown menu.

This acts like a quick reboot. It clears out the frozen code and gives you a completely fresh, clean slate to try again, without losing any of the text answers you have already typed.

##  Activity 2: Creating a Script:
We will create a script that displays **Hello** five times in a single line:

```python
# NOTE: The word Hello is followed by a space and both are placed inside quotation marks. 
# Spaces are also strings in python and will print out along with the word.
# The number 5 is not inside quotation marks.
#
# Click SHIFT+ENTER to run the cell.

print("-----------------------------")
print("Hello " * 5)
print("-----------------------------")
```

Notice that the string **Hello** along with the space is treated like a number and is multiplied 5 times in a line. Also, remember that there is a space after the word? That gives space between the repeated words
